```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('B:/S3/best_imaging_dataset123.csv')
df.head()
```

```python
df['test_date']= pd.to_datetime(df['test_date'],errors='coerce')
df['report_released_at']= pd.to_datetime(df['report_released_at'],errors='coerce')
```

```python
numeric_columns=['cost','result_value','turnaround_time_hours','image_size_mb','hour','no_show','cancellation']
for col in numeric_columns : df[col]=pd.to_numeric(df[col],errors='coerce')
```

```python
df['gender']=df['gender'].str.title()
```

```python
df['is_imaging'] = df['is_imaging'].astype(str).str.lower().map({'true': True, 'false': False})
```

```python
df.fillna({
    'cost': 0,
    'insurance_provider': 'Self Pay',
    'result_flag': 'Unknown',
}, inplace=True)
```

```python
df['year'] = df['test_date'].dt.year
df['month'] = df['test_date'].dt.month
df['day'] = df['test_date'].dt.day
```

```python
df['tat_hours'] = (df['report_released_at'] - df['test_date']).dt.total_seconds() / 3600
```

```python
df.to_csv('B:/S3/best_cleaned_dia.csv', index=False)
df.head()
```
